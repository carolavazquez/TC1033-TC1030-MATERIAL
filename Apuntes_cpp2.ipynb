# **1.0 ¿Qué es un arreglo?**

- Un arreglo en C++ es una estructura de datos que permite almacenar una colección de elementos  
del **mismo tipo bajo un mismo nombre**.
- Estos elementos tienen **el mismo tipo de datos**, y cada elemento se accede a través de un índice usando un solo nombre, facilitando el manejo de cientos o miles de elementos de un arreglo.

# **1.1º Declaración del Arreglo**

La sintaxis básica para declarar un arreglo en C++ es:

```cpp
tipo_de_dato  nombreDelArreglo[size];
```
- Donde:
  - `tipo_de_dato`: Es el tipo de datos de los elementos del arreglo (por ejemplo, int, char, double, etc.).
  - `nombreDelArreglo`: Es el nombre que le das al arreglo.
  - `size`: Es la cantidad de elementos que puede contener el arreglo.

Ejemplo: declaración de un arreglo de enteros con 35 elementos, se usa la siguiente sintaxis:
```cpp
int calificaciones[35];
```

Ejemplo: declaración e inicialización de un arreglo de enteros con 10 elementos, se usa la siguiente sintaxis:
```cpp
                 
int calificaciones2[] = {100, 98, 99, 100, 89, 88, 100, 100, 85, 100}
```

## Índices para accesar un elemento del arreglo
- Para accesar - **obtener el valor , cambiar el valor**- de un elemento del arreglo se utiliza un índice.
- El índice tiene que ser de tipo entero.
- Los índices de un arreglo comienzan desde 0 hasta tamaño - 1.  Por ejemplo:

```cpp
// Asigna el valor 100 al primer elemento del arreglo - índice 0
calificaciones[0] = 100;  
```

- Es importante tener en cuenta que los arreglos en C++ son de tamaño fijo y no pueden cambiar su tamaño una vez que han sido declarados.

### Explicación:

- En el momento que se lleva a cabo la ejecución de la función `inicializa`, el arreglo **`arrA`** modifica directamente el **`arrB`**.
- El hecho de que el arreglo **`arrB`** se modifico es porque todos los arreglos tienen un paso de parámetros por REFERENCIA en forma automática. ✅  
- El parámetro de la función apunta a la misma dirección del argumento correspondiente, y por lo tanto si dentro de la función se modifico, estas modificaciones se hacen el argumento usado al llamar a la función. 💯

`````python
%%writefile inicializaRandom.cpp
#include <iostream>
#include <stdlib.h>     /* srand, rand */
#include <time.h>       /* time */
using namespace std;
void inicializa(int arrA[], int n){
  // Inicializar la semilla random para que sea diferente la secuencia
  srand (time(NULL));
  for (int index = 0; index < n; index++){
    arrA[index] = rand() % 2000 - 1000;
  }
}

void despliega(int arrA[], int n){
  // es ineficiente ---
  for (int index = 0; index < n ; index++){
    cout << arrA[index] << ", ";
    if (index == n - 1){
    cout << arrA[index] ;
   }

  for (int index = 0; index < n - 1; index++){
    cout << arrA[index] << ", ";
  }
  if (n > 0){
    cout << arrA[n - 1] << "]\n";
  }
  else{
    cout << "]\n";
  }
}

int main(){
  // 1º Declaración del arreglo
  int arrB[10];

  // 2º llama a la función que inicializa los elementos del arreglo
  inicializa(arrB,10);
  despliega(arrB, 10);
  return 0;
}
`````

# **1.5 Obtener, calcular la cantidad de elementos de un arreglo en C++**  

 - Usa la función `sizeof` o calcular el tamaño del arreglo dividiendo el tamaño total del arreglo por el tamaño de un elemento.

 - La función `sizeof(arr)` en C++ retorna el tamaño total del arreglo arr en bytes.   
 Este tamaño incluye todos los elementos del arreglo y cualquier posible espacio adicional debido a la alineación de memoria.

 - La sintaxis general para utilizar sizeof es sizeof(var), donde var es la variable para la cual deseas obtener el tamaño en bytes.

- Por ejemplo, si tienes un arreglo de enteros int arr[5], y si `sizeof(int)` es 4 bytes en tu plataforma, entonces `sizeof(arr)` te devolverá el tamaño total del arreglo en bytes.    
-  Si `sizeof(int)`  es 4 bytes, entonces  `sizeof(arr)` será igual a 5 * 4, es decir, 20 bytes.

- Este valor puede variar dependiendo de la arquitectura y el compilador que estés utilizando.



`````python
%%writefile SizeArray.cpp

#include <iostream>
#include <string>
using namespace std;
int main() {
    int size, x;
    int arrA[] = {10, 20, 30, 40, 50};
    string frase = "No dejes para mañana lo que puedas aprender hoy";

    cout << "Ocupación en bytes = " << sizeof(arrA) << endl;
    cout << "Ocupación en bytes de un elemento del arreglo = " <<  sizeof(arrA[0]) << endl;
    cout << "Elementos del arreglo: " << sizeof(arrA) / sizeof(arrA[0]) << endl;

    size = sizeof(arrA) / sizeof(arrA[0]);
    x = arrA[size - 4];
    cout << x << endl;
    cout << "contenido\n";
    cout << arrA[size] << endl;
    cout << arrA[size -1] << endl;
    cout << arrA[size -2] << endl;
    cout << arrA[size -3] << endl;
    cout << arrA[size -4] << endl;

    return 0;
}
`````

# **2. Matrices o arreglos de 2 dimensiones- Declaración**

- Un arreglo de dos dimensiones, también conocido como matriz, es una estructura de datos en programación que consta de renglones y columnas.
- Es como una tabla en la que cada celda contiene un elemento y está identificada por dos índices: uno para el renglón y otro para la columna.

En C++, un arreglo de dos dimensiones se declara de la siguiente manera:

```cpp
tipo de dato nombreMatriz[numRenglones][numColumnas];
```

- `tipo de dato`: Es el tipo de datos que contendrá cada elemento de la matriz (por ejemplo, int, char, double, etc.).
- `nombreMatriz`: Es el nombre que le das al arreglo de dos dimensiones.
- `numRenglones` y `numColumnas`: Son cantidad de renglones y columnas de la matriz.

Por ejemplo, puedes declarar una matriz de enteros 3X3 de la siguiente manera:

```cpp
int matriz[3][3];
```

- Acceso a un elemento específico  se deben utilizar dos índices, uno para la fila o renglón y otro para la columna:

```cpp
matriz[0][0] = 10;  // Asigna el valor 10 a la posicion 0,0 de la matriz
```

Gráficamente:


| renglon | col_0   | col_1   | col_2 |  
| ------- | -----------  | -----------  | ----------- |  
|0| matriz[0][0] | matriz[0][1] | matriz[0][2] |
|1| matriz[1][0] | matriz[1][1] | matriz[1][2] |
|2| matriz[2][0] | matriz[2][1] | matriz[2][2] |


Un arreglo de dos dimensiones se utiliza para representar estructuras de datos como tablas, imágenes bidimensionales, mapeos y muchas otras aplicaciones que requieren organizar datos en forma bidimensional.

# String - métodos


La clase `std::string` en C++ proporciona una variedad de métodos para manipular y trabajar con cadenas de caracteres. A continuación, se presentan algunos de los métodos más comunes junto con una breve explicación de cada uno:

1. **Constructor:**
   - `string()`: Constructor predeterminado que crea una cadena vacía.


2. **Operadores de Asignación y Concatenación:**
   - `operator=()`: Asigna una cadena a otra.
   - `operator+=()`: Concatena una cadena con otra.

3. **Elemento Acceso:**
   - `operator[]`: Devuelve el carácter en la posición especificada.
  

4. **Iteradores:**
   - `begin()`: Devuelve un iterador al primer carácter de la cadena.
   - `end()`: Devuelve un iterador al carácter después del último de la cadena.

5. **Capacidad:**
   - `size() / length()`: Devuelve la longitud de la cadena.
   - `empty()`: Comprueba si la cadena está vacía.
  

6. **Modificación:**
   - `append()`: Agrega caracteres al final de la cadena.
   - `assign()`: Asigna una nueva cadena a la actual.
   - `insert()`: Inserta caracteres en una posición específica.
   - `erase()`: Elimina caracteres de la cadena.
   - `replace()`: Reemplaza una parte de la cadena con otra cadena.

7. **Búsqueda y Subcadenas:**
   - `find()`: Encuentra la primera ocurrencia de una subcadena.
   - `rfind()`: Encuentra la última ocurrencia de una subcadena.
   - `substr()`: Devuelve una subcadena de la cadena.

8. **Manipulación de Caracteres:**
En C++, puedes usar las funciones toupper y tolower de la biblioteca <cctype> para convertir caracteres a mayúsculas y minúsculas, respectivamente.

```cpp
#include <iostream>
#include <cctype>  //  para toupper y tolower


    string str = "Hola, Mundo!";

    str[1] = toupper(str[1]);  // Convierte a mayúsculas
    str[0] = tolower(str[0]);  // Convierte a minúsculas
````

   - `tolower()`: Convierte los caracteres a minúsculas.
   - `toupper()`: Convierte los caracteres a mayúsculas.

9. **Swap:**
   - `swap()`: Intercambia el contenido de dos cadenas.

La clase `std::string` en C++ proporciona numerosos métodos para operar y manipular cadenas de caracteres. Aquí tienes una lista de algunos de los métodos más comunes de la clase `std::string` junto con una breve explicación de cada uno:

1. **`length()` y `size()`:** Devuelven el tamaño (longitud) de la cadena, es decir, el número de caracteres que contiene.

```cpp
string str = "Hola";
int len = str.length(); // len es igual a 4
```

2. **`empty()`:** Devuelve `true` si la cadena está vacía (no contiene caracteres), y `false` en caso contrario.

```cpp
string str = "Hola";
bool isEmpty = str.empty(); // isEmpty es igual a false
```

3. **`clear()`:** Elimina todos los caracteres de la cadena, dejándola vacía.

```cpp
string str = "Hola";
str.clear(); // Ahora str está vacío
```

4. **`append()` y `+=`:** Agregan caracteres o una cadena al final de la cadena actual.

```cpp
string str = "Hola";
str.append(" Mundo"); // str ahora es "Hola Mundo"
str += "!";
```

5. **`insert()`:** Inserta una cadena o caracteres en una posición específica de la cadena actual.

```cpp
string str = "Hola";
str.insert(2, " Mundo"); // str ahora es "Ho Mundo la"
```

6. **`erase()`:** Elimina un rango de caracteres de la cadena.

```cpp
string str = "Hola Mundo";
str.erase(4, 5); // str ahora es "Hola"
```

7. **`replace()`:** Reemplaza una porción de la cadena por otra cadena o caracteres.

```cpp
string str = "Hola Mundo";
str.replace(5, 5, "Amigos"); // str ahora es "Hola Amigos"
```

8. **`substr()`:** Devuelve una subcadena a partir de una posición dada y con una longitud determinada.

```cpp
string str = "Hola Mundo";
string sub = str.substr(5, 5); // sub es "Mundo"
```

9. **`find()` y `rfind()`:** Buscan la primera (o última) aparición de una subcadena en la cadena y devuelven la posición donde se encontró. Si no se encuentra, devuelve `std::string::npos`.

```cpp
string str = "Hola Mundo";
size_t pos = str.find("Mundo"); // pos es igual a 5
```


Estos son algunos de los métodos más utilizados de la clase `std::string` en C++. La clase también incluye otros métodos para realizar tareas más específicas, como la conversión entre tipos de datos y la manipulación de mayúsculas y minúsculas, entre otros.
